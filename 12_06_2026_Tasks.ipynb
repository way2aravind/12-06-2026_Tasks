Task 1: Square Numbers
Question

Create a list of squares from 1 to 10 using list comprehension.

```python
squaresList = [i**2 for i in range(1,11)]
```

Task 2: Even Numbers
Question

Create a list containing only even numbers from 1 to 20.

```python
even_numberList = [i for i in range(1,21) if i%2 ==0]
```

```python
even_numberList
```

Task 3: Convert Names to Uppercase
Question

Convert all names in a list to uppercase using list comprehension.

```python
Teams = ["Argentina","Portugal","France","Croatia","Netharlands"]
```

```python
uppercaseTeams = [i.upper() for i in Teams]
```

```python
uppercaseTeams
```

Task 4: Find Length of Each Word
Question

Create a list containing the length of each word.

```python
players = ["Messi","Emi","Lautaro","Julian","Enzo","Paredes",]
```

```python
wordCount = [len(i) for i in players]
```

```python
wordCount
```

Task 5: Extract Positive Numbers
Question

Given a list of numbers, create a new list containing only positive numbers.

nums = [-5, 10, -2, 8, 0, 15, -7]

```python
nums = [-5, 10, -2, 8, 0, 15, -7]
```

```python
positiveNums = [i for i in nums if i>0]
```

```python
positiveNums
```
